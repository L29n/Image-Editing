Given an alpha value all the images are blended together. The alpha value determines the weight of both of the images when they are blended. Higher alpha values means the blended result is more heavily impacted by image2, and lower alpha values means the blended result is more heavily impacted by image1.

```python
def blend(image1, image2, alpha):
    blended = alpha * image1 + (1 - alpha) * image2    
    blended = blended.astype(int) #int cast everything for io.imshow()
    
    return blended

alpha = 0.8
blend0to1 = blend(image0, image1, alpha)
blend0to2 = blend(blend0to1, image2, alpha)
blend0to3 = blend(blend0to2, image3, alpha)
blend0to4 = blend(blend0to3, image4, alpha)
blend0to5 = blend(blend0to4, image5, alpha)
blend0to6 = blend(blend0to5, image6, alpha)
blend0to7 = blend(blend0to6, image7, alpha)
blend0to8 = blend(blend0to7, image8, alpha)
blend0to9 = blend(blend0to8, image9, alpha)
blend0to10 = blend(blend0to9, image10, alpha)

```
